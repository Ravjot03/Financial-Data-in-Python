```
import pandas as pd
import numpy as np
from datetime import date
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

## pandas_datareader

pandas_datareader

*   Easy access to various financial internet data sources.
*   Little code needed to import a `pandas` `DataFrame`.
*   Available sources include - 
       *   Federal Reserve
       *   World Bank, OECD, Eurostat
       *   OANDA

```
from pandas_datareader.data import DataReader
```

## FRED
Federal Reserve Economic Data

500,000 series covering a range of categories -
  *  Economic growth & employment
  *  Monetary & fiscal policy
  *  Demographics, industries, commodity prices
  *  Daily, monthly, annual frequencies

```
#Duration of data we are importing
start = date(2000, 1, 1)
end = date(2020, 12, 31)
```

```
series_code = 'UNRATE'   #Series Code for importing UNEMPLOYMENT data from FRED Website
data_source = 'fred'     #Data source

#Importing the UNEMPLOYMENT Data from FRED Website
unemp = DataReader(series_code, data_source, start, end)
unemp.info()
```

```
#Viewing first 5 rows of the dataset
unemp.head()
```

```
#Summary statistics of the dataset
unemp.describe()
```

```
#Shape of the dataset
unemp.shape
```

```
#Maximum Unemployment rate index
unemp.loc[unemp.UNRATE.idxmax()]
```

```
#Minimum Unemployment rate index
unemp.loc[unemp.UNRATE.idxmin()]
```

```
#Plotting the unemployment data
unemp.plot()
plt.show()
```

```
#Adding Mean and Median values to the plot
fig, ax = plt.subplots()
unemp.plot(ax=ax)
ax.axhline(unemp.UNRATE.mean(), color='green', ls='--',label='UNRATE Average')
ax.axhline(unemp.UNRATE.median(), color='red', ls='-.',label='UNRATE Median')
plt.legend()
plt.show()
```

## Naver Finance Data

Naver Finance provides Korean stock market (KOSPI, KOSDAQ) historical data.

```
#Changing the style of seaborn plots
sns.set_style('darkgrid')
```

Importing data of Quater-4 0f 2020.

```
start = date(2020, 10, 1) #start date 01 October, 2020
end = date(2020, 12, 31)  #end date 31 December, 2020
```

Importing the stock market data of **Kia Motors** from **Naver Finance website**, and storing the DataFrame into `kia_motors`.

```
kia_motors = DataReader('000270', 'naver', start, end)
kia_motors.info()
```

Converting all the columns to `float` dtype.

```
kia_motors.Close = kia_motors.Close.astype(float)
kia_motors.Open = kia_motors.Open.astype(float)
kia_motors.Low = kia_motors.Low.astype(float)
kia_motors.High = kia_motors.High.astype(float)
kia_motors.Volume = kia_motors.Volume.astype(float)
```

Updated data

```
kia_motors.info()
```

Importing the stock market data of **Hyundai Motor Company** from **Naver Finance website**, and storing the DataFrame into `hyundai`.

```
hyundai = DataReader('005380', 'naver', start, end)
hyundai.info()
```

Converting all the columns to `float` dtype.

```
hyundai.Close = hyundai.Close.astype(float)
hyundai.Open = hyundai.Open.astype(float)
hyundai.Low = hyundai.Low.astype(float)
hyundai.High = hyundai.High.astype(float)
hyundai.Volume = hyundai.Volume.astype(float)
```

Updated data

```
hyundai.info()
```

First 5 rows of `kia_motors` DataFrame.

```
kia_motors.head()
```

First 5 rows of `hyundai` DataFrame.

```
hyundai.head()
```

Creating a list of few column names `Open`, `Close`, `High` and `Low`.

```
stocks = ['Open','Close','High','Low']
```

Plotting the Kia Motors Stock Prices over the last quater of 2020.

```
kia_motors[stocks].plot()
plt.title("Kia Motors Stock Prices over 4th Quater 2020")
plt.show()
```

Plotting the Hyundai Stock Prices over the last quater of 2020.

```
hyundai[stocks].plot()
plt.title("Hyundai Stock Prices over 4th Quater 2020")
plt.show()
```

Kia Motors vs Hyundai Stock Prices

```
fig, ax= plt.subplots(figsize=[12,12])
kia_motors[stocks].add_suffix("_kia_motors").plot(ax=ax)
hyundai[stocks].add_suffix("_hyundai").plot(ax=ax)
plt.title("Kia Motors vs Hyundai Motor Company Stock Prices in Q-4 2020")
plt.show()
```

`Open` Stock Prices of both of the companies

```
fig, ax= plt.subplots(figsize=[8,4])
kia_motors['Open'].plot(ax=ax, label='Kia')
hyundai['Open'].plot(ax=ax, label='Hyundai')
plt.legend()
plt.title("Kia Motors vs Hyundai Daily Open Stock Prices over Q-4 2020")
plt.show()
```

Kia Motors vs Hyundai Motor Compay Stocks traded (Volume) in Q-4 2020

```
fig, ax= plt.subplots(figsize=[8,4])
kia_motors['Volume'].plot(ax=ax, label='Kia Stocks')
hyundai['Volume'].plot(ax=ax, label='Hyundai Stocks')
plt.title("Kia vs Hyundai Stocks traded in Q-4 2020")
plt.legend()
plt.show()
```
